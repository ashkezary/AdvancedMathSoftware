```python
inters = {}
for x in interests:
    if(x[1] in inters):
        inters[x[1]] = inters[x[1]]+1
    else:
        inters[x[1]] = 1
inters = dict(sorted(inters.items(), key=lambda item: item[1]))
print(inters)
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
plt.xticks(rotation='vertical')
plt.bar(inters.keys(), inters.values(), color='g')
```

2nd. line graph

```python
import numpy as np
x = np.array([i for i in range(33)])
y = np.fromiter(inters.values(), dtype=int)
my_xticks = inters.keys()
plt.xticks(x, my_xticks,rotation='vertical')
plt.plot(x, y)
plt.show()
```

<b>3rd. pie chart</b>

```python
# Data to plot
labels = []
sizes = []

for x, y in inters.items():
    labels.append(x)
    sizes.append(y)

# Plot
plt.pie(sizes, labels=labels,labeldistance=1.0,rotatelabels =True,startangle = 10)

plt.axis('equal')
plt.show()
```
